# Interactive Colour Scale Bar

Run two hidden cells below, then use interactive widgets to design colour bar.

```python
from ipyfilechooser import FileChooser
from ipywidgets import (BoundedFloatText, Dropdown, Checkbox, Text, Label,
                        BoundedIntText, Button, Layout, HBox, VBox, HTML,
                        RadioButtons, Output)
from IPython.display import display, HTML as htm

from cbar import drawCBar, is_mpl_available, cbar_usage
if is_mpl_available:
    import matplotlib.pyplot as plt
```

```python
myCB = None
kw = {}


def changed_dir(chooser):
    """Update fcsvg."""
    #
    if fcsvg.selected_filename is None:
        fcsvg.default_path = chooser.selected_path
    cbaro.outputs = []


def createCB(btn):
    '''Create, save and display the colour bar.'''
    #
    global myCB, kw
    #
    cbaro.outputs = []
    qml = fcqml.value
    if qml is None:
        with cbaro:
            print('You must provide a .qml file!!!')
        return

    svg = fcsvg.value
    kw['width'] = wwdth.value
    kw['height'] = whght.value
    kw['drawedges'] = wedg.value
    kw['inverted'] = winv.value
    kw['orientation'] = wori.options[wori.value-1][0].lower()
    kw['spacing'] = wspc.options[wspc.value-1][0].lower()
    kw['title'] = wtitl.value.replace('\\n','\n')
    kw['titlefontname'] = wtifn.value if wtifn1.value-1 == 0 else wtifn1.options[wtifn1.value-1][0]
    kw['titlefontsize'] = wtsz.value if wtsz1.value-1 == 0 else wtsz1.options[wtsz1.value-1][0]
    kw['titlefontweight'] = wtfw.value if wtfw1.value-1 == 0 else wtfw1.options[wtfw1.value-1][0]
    kw['titlecolo'] = wtcol.value
    kw['tickfontsize'] = wisz.value if wisz1.value-1 == 0 else wisz1.options[wisz1.value-1][0]
    kw['tickcolo'] = wicol.value
    kw['tickstep'] = wistp.value
    kw['decimal'] = wicml.value
    kw['ticklength'] = wilth.value
    kw['bottom'] = wbot.value
    kw['top'] = wtop.value
    kw['left'] = wlft.value
    kw['right'] = wrgt.value
    kw['tweak_end_replace'] = True if wtw.value == wtw.options[1] else False
    kw['tweak_end_add'] = True if wtw.value == wtw.options[2] else False
    kw['verbose'] = wverb.value
    #
    myCB = drawCBar(qml, svg, **kw)
    b = myCB.createCBar()
    if b:
        with cbaro:
            plt.show()
            print('\nCreated svg file:', myCB.the_svg)
            hh = '<html><body><p><a href="file:///%s">Shift-Right-Click -&gt; Open link in a new tab</a></p></body></html>' % myCB.the_svg
            display(htm(hh))
            if wverb.value:
                print(qml, '\n', svg, '\n', kw, '\n', '='*90)
                print(myCB.L_Verbose)
    else:
        with cbaro:
            print('An ERROR occurred:', myCB.err)
    myCB.close()


# Define user interface
wall = Layout(width='500px')
w500 = Layout(width='150px')
w300 = Layout(width='50px')
w200 = Layout(width='100px')
w075 = Layout(width='75px')
w050 = Layout(width='50px')
w020 = Layout(width='25px')

# colour bar
wwdth = BoundedFloatText(value=15, min=0.2, max=50, step=5, disabled=False, layout=w075)
whght = BoundedFloatText(value=0.75, min=0.1, max=15, step=0.5, disabled=False, layout=w075)
wedg = Checkbox(value=False, disabled=False, indent=False, layout=w020)
wori = Dropdown(options=[('Horizontal', 1), ('Vertical', 2)], value=1, layout=w200)
wspc = Dropdown(options=[('Uniform', 1), ('Proportional', 2)], value=1, layout=w200)
winv = Checkbox(value=False, disabled=False, indent=False, layout=w020)

# Title
wtitl = Text(value='', placeholder='Title', disabled=False, layout=wall)
wtifn1 = Dropdown(options=[('', 1), ('serif', 2), ('sans-serif', 3), ('cursive', 4),
                           ('fantasy', 5), ('monospace', 6)], value=3, layout=w200)
wtifn = Text(value='', placeholder='Title font name', disabled=False, layout=w500)
wtsz1 = Dropdown(options=[('', 1), ('xx-small', 2), ('x-small', 3), ('small', 4), ('medium', 5), ('large', 6),
                          ('x-large', 7), ('xx-large', 8)], value=1, layout=w200)
wtsz = BoundedIntText(value=28, min=1, max=192, step=1, disabled=False, layout=w075)
wtfw1 = Dropdown(options=[('', 1), ('ultralight', 2), ('light', 3), ('normal', 4),
                          ('demibold', 5), ('bold', 6), ('extra bold', 7)], value=6, layout=w200)
wtfw = BoundedIntText(value=0, min=0, max=1000, step=10, disabled=False, layout=w075)
wtcol = Text(value='k', placeholder='colourcode, named colour or #xxxxxx', disabled=False, layout=w200)

# Ticks labels
wisz1 = Dropdown(options=[('', 1), ('xx-small', 2), ('x-small', 3), ('small', 4), ('medium', 5), ('large', 6),
                          ('x-large', 7), ('xx-large', 8)], value=1, layout=w200)
wisz = BoundedIntText(value=16, min=1, max=192, step=1, disabled=False, layout=w075)
wicol = Text(value='k', placeholder='colour code, named colour or #xxxxxx', disabled=False, layout=w200)
wistp = BoundedIntText(value=-5, min=-5, max=10000, step=5, disabled=False, layout=w075)
wicml = BoundedIntText(value=3, min=0, max=15, step=1, disabled=False, layout=w300)
wilth = BoundedFloatText(value=8, min=0.2, max=20, step=0.1, disabled=False, layout=w075)
# Ticks/Labels placement
wbot = Checkbox(value=True, disabled=False, indent=False, layout=w300)
wlft = Checkbox(value=True, disabled=False, indent=False, layout=w300)
wtop = Checkbox(value=False, disabled=False, indent=False, layout=w020)
wrgt = Checkbox(value=False, disabled=False, indent=False, layout=w020)

#Tweak
wtw = RadioButtons(value='Replace last tick', options=['None', 'Replace last tick', 'Add end tick'], description='Tweak ticks')
wverb = Checkbox(value=False, disabled=False, indent=False, layout=w200)

whlp = HTML(cbar_usage, layout=Layout(width='500px', border='solid 1px', padding='0px 0px 0px 15px', margin='5px'))
wbtngo = Button(description='Create colour bar', disabled=False, button_style='', tooltip='Create and save the colour bar')
fcqml = FileChooser('')
fcqml.filter_pattern = ['*.qml', '*.QML']
fcsvg = FileChooser('')
fcsvg.filter_pattern = ['*.svg', '*.SVG']
cbaro = Output()

# Setup callbacks
wbtngo.on_click(createCB)
fcqml.register_callback(changed_dir)


fcqml.default_path = r'D:\Dev\bcc_QGIS_Plugins\git\data\qml-examples'


# Layout
lfcqml = HBox([Label('QGIS V3.x qml:', layout=w200), fcqml])
lfcsvg = HBox([Label('Output cbar svg:', layout=w200), fcsvg])
# colour bar
lwwdth = HBox([wwdth, Label(' : Length', layout=w500)])
lwhght = HBox([whght, Label(' : width', layout=w500)])
lwedg = HBox([wedg, Label(' : Draw colour edges', layout=w500)])
lwinv = HBox([winv, Label(' : Reverse colour bar', layout=w500)])
lwori = HBox([wori, Label(' : Orientation', layout=w500)])
lwspc = HBox([wspc, Label(' : Colour spacing', layout=w500)])

# Title
lwtitl = HBox([Label('Title:', layout=w050), wtitl])
lwtifn = HBox([wtifn, wtifn1, Label(' : Title font name', layout=w500)])
lwtsz = HBox([wtsz, wtsz1, Label(' : Title font size', layout=w500)])
lwtfw = HBox([wtfw, wtfw1, Label(': Title font weight', layout=w500)])
lwtcol = HBox([wtcol, Label(' : Title colour', layout=w500)])

# Ticks labels
lwisz = HBox([wisz, wisz1, Label(' : Tick label font size', layout=w500)])
lwicol = HBox([wicol, Label(' : Tick label colour', layout=w500)])
lwistp = HBox([wistp, Label(' : Tick separation', layout=w500)])
lwicml = HBox([wicml, Label(' : Decimal places', layout=w500)])
lwilth = HBox([wilth, Label(' : Tick length', layout=w500)])
# Ticks/Labels placement
lwbot = HBox([Label('Ticks below:', layout=w200), wbot])
lwlft = HBox([Label('Ticks on left:', layout=w200), wlft])
lwtop = HBox([wtop, Label(' : Ticks on top', layout=w200)])
lwrgt = HBox([wrgt, Label(' : Ticks on right', layout=w200)])

#Tweak
lwverb = HBox([Label('Verbose:', layout=w200), wverb])

# Setup UI
items_layout = Layout(width='auto')     # override the default width of the button to 'auto' to let the button grow
vert_layout = Layout(
    display='flex',
    flex_flow='column',
    align_items='stretch',
    border=None,
    width='100%'
)
hori_layout = Layout(
    display='flex',
    flex_flow='row',
    align_items='stretch',
    width='100%',
    border=None,
    padding='3px',
    margin='auto'
)

wcbartitre = HTML('<h1><font color="#090">Create a Colour Scale Bar from QGIS .qml file</font></h1>', Layout=wall)
wl_cb1 = HBox([lwwdth, lwhght])
# wl_cb2 = HBox([lwori, lwspc])
wl_cb2 = HBox([lwori])
wl_cb3 = HBox([lwedg, lwinv])
wl_ti0 = HBox([lwtitl])
wl_ti00 = HBox([lwtifn])
wl_ti1 = HBox([lwtsz])
wl_ti2 = HBox([lwtfw])
wl_ti3 = HBox([lwtcol])
wl_ii1 = HBox([lwisz])
wl_ii2 = HBox([lwicol, lwistp])
wl_ii3 = HBox([lwicml, lwilth])
wl_ii4 = HBox([lwbot, lwtop])
wl_ii5 = HBox([lwlft, lwrgt])
wl_twk = HBox([wtw, lwverb])

box = VBox([wcbartitre, lfcqml, lfcsvg,
            HTML('&nbsp;<br /><b><font color="#006">Colour bar parameters</font></b>', Layout=wall), wl_cb1, wl_cb2, wl_cb3,
            HTML('&nbsp;<br /><b><font color="#006">Title parameters</font></b>', Layout=wall), wl_ti0, wl_ti00, wl_ti1, wl_ti2, wl_ti3, 
            HTML('&nbsp;<br /><b><font color="#006">Ticks/Labels parameters</font></b>', Layout=wall), wl_ii1, wl_ii2, wl_ii3, wl_ii4, wl_ii5,
            HTML('&nbsp;<br /><b><font color="#006">Tweaking parameters</font></b>', Layout=wall), wl_twk,
            Label(' ', Layout=wall),
            wbtngo,
            Label(' ', Layout=wall),
            cbaro], layout=Layout(width='700px', border='solid 1px', padding='5px', margin='5px'))

ui = HBox([box, whlp])
ui
```
